```python
import arcpy
import ftplib
```

```python
# Connect to FTP directory for Ramsey County Lidar LAZ files
ftp = ftplib.FTP('ftp.lmic.state.mn.us')
ftp.login()
ftp.cwd('/pub/data/elevation/lidar/county/ramsey/laz')
```

```python
# Downloads all files from FTP directory to LAS files project folder on local device

filenames = ftp.nlst()

for file in filenames:
    print("Downloading..." + file)
    ftp.retrbinary("RETR " + file ,open("C:/Users/Holly Leaf/GIS5571/Lab2/Lab2 part1/LAS files/" + file, 'wb').write)

ftp.quit()
print(str(len(filenames)) + " files have been downloaded.")
```

```python
# Convert LAZ files to LAS using Arc

arcpy.conversion.ConvertLas(r"C:\Users\Holly Leaf\GIS5571\Lab2\Lab2 part1\LAS files", r"C:\Users\Holly Leaf\GIS5571\Lab2\Lab2 part1\LAS files\converted las", "SAME_AS_INPUT", '', "NO_COMPRESSION", "REARRANGE_POINTS", None)
```

```python
# Convert LAS dataset to raster (DEM)

arcpy.conversion.LasDatasetToRaster("3542-29-27.las", r"c:\users\holly leaf\gis5571\lab2\lab2 part1\lab2 part1.gdb\c35422927_la", "ELEVATION", "BINNING AVERAGE LINEAR", "FLOAT", "CELLSIZE", 10, 1)
```

```python
#Saves DEM to disk as TIFF

arcpy.management.CopyRaster("c35422927_la", r"C:\Users\Holly Leaf\GIS5571\Lab2\Lab2 part1\c35422927_la", '', None, "3.4e+38", "NONE", "NONE", '', "NONE", "NONE", "TIFF", "NONE", "CURRENT_SLICE", "NO_TRANSPOSE")
```

```python
# Convert LAS dataset to TIN

arcpy.ddd.LasDatasetToTin("3542-29-27.las", r"C:\Users\Holly Leaf\GIS5571\Lab2\Lab2 part1\c3542-29-27_LasDatasetToTin-thinned75", "RANDOM", "PERCENT", 75, 20000000, 1, "CLIP")
```
